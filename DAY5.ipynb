```python
import pandas as pd
employees = {
"Emp_ID" : [1001,1002,1003],
"Name" : ["Nithish","siva","karthi"],
"Department" : ["IT","HR","IT"],
"Salary" : [10000,15000,18000],
"Experience" : [3,4,3]
}
df = pd.DataFrame(employees)
print("Employee Dataset")
print(df)
print("\nAverage Salary")
print(df["Salary"].mean())
print("\nHighest Salary")
print(df["Salary"].max())
```
