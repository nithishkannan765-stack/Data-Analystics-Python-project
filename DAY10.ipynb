```python
import matplotlib.pyplot as plt
students = ["a","b","c","d","e"]
Marks = [75,45,35,65,100]
plt.plot(students, Marks)
plt.title("students Marks")
plt.xlabel("students")
plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["a","b","c","d","e"]
Marks = [75,45,35,65,100]
plt.pie(Marks, labels=students, autopct='%1.1f%%', startangle=90)
plt.title("students Marks")
# plt.xlabel("students") # Not applicable for pie chart
# plt.ylabel("Marks") # Not applicable for pie chart

plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["a","b","c","d","e"]
Marks = [75,45,35,65,100]
plt.bar(students, Marks)
plt.title("students Marks")
plt.xlabel("students")
plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["a","b","c","d","e"]
Marks = [75,45,35,65,100]
plt.hist(Marks, bins=5, edgecolor='black') # Plotting histogram of Marks with 5 bins
plt.title("Students Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["a","b","c","d","e"]
Marks = [75,45,35,65,100]
plt.boxplot(Marks, labels=["Student Marks"]) # Pass numerical Marks and label the box plot
plt.title("Students Marks Distribution")
plt.xlabel("Categories") # Changed to a more general label for a single box plot
plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["a","b","c","d","e"]
Marks = [75,45,35,65,100]
plt.stackplot(students, Marks)
plt.title("students Marks")
plt.xlabel("students")
plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["a","b","c","d","e"]
Marks = [75,45,35,65,100]
plt.stem(students, Marks)
plt.title("students Marks")
plt.xlabel("students")
plt.ylabel("Marks")

plt.show()
```
