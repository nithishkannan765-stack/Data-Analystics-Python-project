```python
import pandas as pd
data = {
"Name": ["Name1","Name1","Name3","Name4"],
"Product": ["laptop","mobile","tablet","biscuit"],
"Salary": [10550,12000,12000,45000]
}
df = pd.DataFrame(data)
Sorted_df = df.sort_values("Product")
print(Sorted_df)
```

```python
print(df["Salary"].min())
```

```python
print(df["Salary"].max())
```

```python
print(df["Salary"].count())
```

```python
region_sales = df.groupby("Name")["Salary"].sum()
print(region_sales)
```
